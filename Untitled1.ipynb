 # Matplotlib Charts 

## 1.Line Chart - shows trends or changes over time 

```python
import matplotlib.pyplot as plt 
x = [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis") 
plt.show()
```

## 2.Bar Chart- compare values among different  categories 

```python
import matplotlib.pyplot as plt 
Categories = ["A","B","C","D"]
values = [25,40,30,35]

plt.bar(Categories, values)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("values") 
plt.show()
```

## 3.Pie Chart- shows the proportion or percentage of each categories 

```python
import matplotlib.pyplot as plt

sizes = [40,30,20,10]
labels = ["A","B","C","D"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie chart")
plt.show()
```

## 4.Scatter plot-shows the relationship between two numerical variables 

```python
import matplotlib.pyplot as plt 
x = [1,2,3,4,5]
y = [2,4,5,4,6]

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis") 
plt.show()
```

## 5.Histogram- Display the frequency distribution of data 

```python
import matplotlib.pyplot as plt 
data = [12, 15, 18, 18, 20, 22, 24, 25, 25, 30]

plt.hist(data,bins=5)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency") 
plt.show()
```

## 6.Heat map- highligths the patterns using color intensity 

```python
import matplotlib.pyplot as plt 
data = [
    [5,8,2],
    [7,3,6],
    [4,9,1]
]
plt.imshow(data,cmap='hot',interpolation='nearest')
plt.colorbar()
plt.title("Heat Map")
plt.show()
```

## 7. Box plot-  shows the distribution of data, median,quartiles and outliers

```python
import matplotlib.pyplot as plt 

data = [10,20,25,30,35,40,45]

plt.boxplot(data)
plt.title("Box plot")
plt.show()
```

## 8.Area Chart- shows cumulative values over time 

```python
import matplotlib.pyplot as plt 
x = [1,2,3,4]
y = [5,10,15,20]

plt.fill_between(x,y)
plt.title("Area Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis") 
plt.show()
```

## 9.Stem Plot- Display discrete points

```python
import matplotlib.pyplot as plt 
x = [1,2,3,4]
y = [10,15,8,20]

plt.stem(x,y)
plt.title("Stem Plot")
plt.show()
```

## 10.Horizontal bar chart- compare catogries horizontally 

```python
import matplotlib.pyplot as plt 

categories = ["A","B","C"]
values = [10,20,15]

plt.barh(categories,values)
plt.title("Horizontal Bar Chart")
plt.show()
         
```

## 11.Error Bar Chart- display data withh error or uncertainty 

```python
import matplotlib.pyplot as plt 
x = [1,2,3]
y = [10,15,20]
error = [1,2,1]

plt.errorbar(x,y,yerr=error)
plt.title("Error Bar Chart")
plt.show()
```

## 12.Map- visualize geographical data like cities ,population 

```python
import matplotlib.pyplot as plt 

longitude = 77.1025
latitude = 28.7041
plt.scatter(longitude, latitude, color='red')
plt.title("Location on map")
plt.xlabel("Longitude")
plt.ylabel("Lattitude")
plt.show()
```

## 13.Stack plot- shows the cumulative contribution of multiple categories over time

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [2, 3, 4, 5]
y2 = [1, 2, 1, 2]
y3 = [3, 1, 2, 1]

plt.stackplot(x, y1, y2, y3, labels=['A', 'B', 'C'])
plt.legend()
plt.show()

```

## 14.Hexbin plot- Displays the density of large datasets

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)
y = np.random.randn(1000)

plt.hexbin(x, y, gridsize=20, cmap='Blues')
plt.colorbar(label='Count')
plt.title("Hexbin Plot")
plt.show()
```

## 15.Step plot- chart where data changes in horizontal and vertical steps instead of a continuous line

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 10, 20, 20, 30]

plt.step(x, y)

plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Step Plot Example")

plt.grid(True)
plt.show()hart where data changes in horizontal and vertical steps instead of a continuous line
```

## 16.Error plot- shows data points along with their uncertainty, variation, or measurement error

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [20, 25, 22, 30]
error = [2, 3, 1, 2]

plt.errorbar(
    x,
    y,
    yerr=error,
    fmt='o',
    capsize=5,
    color='red'
)

plt.show()
```

## 17.3D surface plot- displays a three-dimensional surface 

X-axis → first variable
Y-axis → second variable
Z-axis → calculated value/heig

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)

X, Y = np.meshgrid(x, y)

Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.plot_surface(
    X,
    Y,
    Z,
    cmap="coolwarm"
)

plt.show()
```

## 18. Contour plot-Represents 3D information in 2D using lines

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)

X,Y = np.meshgrid(x,y)

Z = X**2 + Y**2

plt.contour(X,Y,Z)

plt.show()
```

## 19.Polar plot-Used for circular/angular data

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0,2*np.pi,100)
r = np.sin(3*theta)

plt.subplot(projection="polar")
plt.plot(theta,r)

plt.show()
```

## 20.Violin plot-combines features of a box plot and a density plot

```python
import matplotlib.pyplot as plt

data = [
    [10, 12, 15, 18, 20, 22],
    [15, 18, 20, 25, 30, 35],
    [5, 8, 12, 15, 18, 20]
]

plt.violinplot(data)

plt.title("Basic Violin Plot")
plt.xlabel("Groups")
plt.ylabel("Values")

plt.show()
```
